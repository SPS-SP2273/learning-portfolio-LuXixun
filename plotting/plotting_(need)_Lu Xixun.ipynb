# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

### Things to note

|**Long form**|**Abbreviation**|
|:--|:--|
|`color`|`c`|
|`linestyle`|`ls`|
|`linewidth`|`lw`|

* both the following lines produce the same result:

```python
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
```

```python
plt.plot(x, y, c='red', ls='dashed', lw=2)
```

* The plotting functions usually have default values for the styling parameters. So, if you wish, you can keep it simple and plot just using:

```python
plt.plot(x, y_1, y_2)
```

* Split the code into seperate lines to improve readability:

```python
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
```

```python
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
```

* The order of how you specify the keyword arguments (color, linewidth, â€¦) does not matter.

# 2 Better with NumPy

Often, it is easier to use NumPy arrays instead of Python lists. Thus, we can first convert the Python lists into NumPy arrays and then redo the plot in the previous step.

```python
import numpy as np
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

 our plot does not look good because 
sin(x) lies between Â±1, but x^2 has no such bounds. One way to fix this is to add another y-axis that shares the same x-axis.

## We need another axis!

use `twinx()` to add another y axis:

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

### Things to note

* We now have two legend() calls, one for each axis.

* our plot still does not look good because we have only a few points. Letâ€™s use np.linspace to fix this with:

```python
x = np.linspace(0, 6, 100)
```

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 3 Saving to disc

To export the plot, we need to save it to the disc. 
To do this, we can do the following:
* Use the function `savefig()`.
* Specify the format with the extension (e.g., `filename.pdf`) of the file name
* Specify resolution by use dots-per-inch (`dpi`).

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

When you run this code, you will find the file saved in the same directory (folder) as the one in your notebook lives in. If you want it saved elsewhere, specify the path in more detail. For example: 

```python
plt.savefig('~/Desktop/simple-plot.png', dpi=150)
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')  # add horizontal line at the zero value
```

`hlines()` function needs a y-value and starting and ending values of x.

## 4.2 Styles

Matpoltlib templates: [here](http://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html)

```python
plt.style.use('fivethirtyeight') #put at the top
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')  # add horizontal line at the zero value
plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
