**Exercise 1 (Fill in the blanks)** Fill in the blanks(indicated with a ?), so the code below will produce the plot shown.

```python
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle= 'dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label= 'Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

**Exercise 2 (A simple mathematical plot)**

```python
import numpy as np


x_values = np.arange(0, 50.1, 0.1)

y_values = np.exp(-x_values/10) * np.sin(x_values)

plt.ylim(-1, 1)

plt.plot(x_values, y_values, color = 'steelblue')
plt.xlabel('x', color = 'darkorange', fontsize = 15)
plt.ylabel('y', color = 'darkorange', fontsize = 15, rotation = 0)
plt.grid(alpha=.25)
plt.suptitle(r'$y = e^{-x/10}\sin(x)$', fontsize = 20)
plt.tight_layout()
```
